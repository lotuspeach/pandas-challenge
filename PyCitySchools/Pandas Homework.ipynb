```python
import pandas as pd
```

```python
#path
student_path = "Resources/students_complete.csv"
school_path = "Resources/schools_complete.csv"
```

```python
#read in
student_data_df = pd.read_csv(student_path)
school_data_df = pd.read_csv(school_path)
```

```python
student_data_df.head()
```

```python
school_data_df = school_data_df.rename(columns={"size" :"Total_Students_Per_School",
                                               "budget": "Budget_Per_School",
                                               "type": "School_Type"})
school_data_df.head()
```

```python
total_budget = school_data_df["Budget_Per_School"].sum()
total_budget
```

```python
total_schools = school_data_df["school_name"].count()
total_schools
```

```python
total_students = school_data_df["Total_Students_Per_School"].sum()
total_students
```

```python
average_mathscore = student_data_df["math_score"].mean()
average_mathscore
```

```python
average_readingscore = student_data_df["reading_score"].mean()
average_readingscore
```

```python
passing_mathstudents = student_data_df.loc[student_data_df["math_score"] > 70] 
percent_passing_math = len(passing_mathstudents)/total_students * 100
percent_passing_math
```

```python
passing_reading_students = student_data_df.loc[student_data_df["reading_score"] > 70] 
percent_passing_reading = len(passing_reading_students)/total_students * 100
percent_passing_reading
```

```python
percent_overall_passing = (percent_passing_math + percent_passing_reading)/2
percent_overall_passing
```

```python
District_Summary_df = pd.DataFrame({"Total schools":total_schools,
                                "Total students": total_students,
                                "Total Budget": total_budget,
                                "Average Math Score":average_mathscore,
                                "Average Reading Score": average_readingscore,
                                "Percent Passing Math":percent_passing_math,
                                "Pecent Passing Reading":percent_passing_reading,
                                "Percent Overall Passing": [percent_overall_passing]})
District_Summary_df
```

```python
grade_nine = student_data_df.loc[student_data_df["grade"]=="9th"].groupby("school_name")
grade_nine_math = grade_nine["math_score"].mean()

grade_ten = student_data_df.loc[student_data_df["grade"]=="10th"].groupby("school_name")
grade_ten_math = grade_ten["math_score"].mean()

grade_eleven = student_data_df.loc[student_data_df["grade"]=="11th"].groupby("school_name")
grade_eleven_math = grade_eleven["math_score"].mean()

grade_twelve = student_data_df.loc[student_data_df["grade"]=="12th"].groupby("school_name")
grade_twelve_math = grade_twelve["math_score"].mean()
```

```python
reading_scores_bygrade_df = pd.DataFrame({"9th": grade_nine_read,
                                      "10th": grade_ten_read,
                                      "11th": grade_eleven_read,
                                      "12th": grade_twelve_read})
reading_scores_bygrade_df.head()
```

```python
#Scores by School Spending
bins_scores = [0, 590, 625, 655, 680]
spending = ["0 to $590", "$590 to $625", "$625 to $655", "$655 to $680"]
School_Summary_df["spending (per student)"] = pd.cut(School_Summary_df["Per Student Budget"], bins_scores, labels=spending)
spending_summary = School_Summary_df.groupby("spending (per student)")
spending_summary = spending_summary[["Average Math Score", "Average Reading Score","Percent Passing Math", "Percent Passing Reading", "Percent Overall Passing"]]
```
